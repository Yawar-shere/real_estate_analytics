```
# File location and type
file_location = "/FileStore/tables/data-1.json"
file_type = "json"

df1 = spark.read.option("multiline", "True").json(file_location)
display(df1)
```

```
from pyspark.sql.functions import explode

ads = df1.select(explode("ads").alias("ads"))
display(ads)
```

```
from pyspark.sql.functions import col

all_ads = ads.select(
    col("ads.ad#").alias("adID")
    , col("ads.Title").alias("Title")
    , col("ads.Location").alias("Location")
    , col("ads.Priority").alias("AdPriority")
    , col("ads.Price").alias("Cost(Rs)")
    , col("ads.`No. of pictures`").alias("Pictures") # use ` to ignore dot operator
    , col("ads.Means").alias("Communication")
    , col("ads.BedBathandArea").alias("BedBathandArea")
    , col("ads.Added").alias("AdDuration")
    , col("ads.URL").alias("AdURL")
    )
display(all_ads)
```

```
from pyspark.sql.functions import split, col, when, array_contains, ceil

df2 = all_ads.withColumn("Cost(Rs)", split(col("Cost(Rs)"), " ")
                        )
display(df2)
```

```
df3 = df2.withColumn(
    "Cost(Rs)", when((array_contains(col("Cost(Rs)"), "Lakh")), (col("Cost(Rs)")[0].cast("float")) * 100000)
    .when((array_contains(col("Cost(Rs)"), "Crore")), (col("Cost(Rs)")[0].cast("float")) * 10000000)
                    )

display(df3)
```

```
df4 = df3.withColumn(
    "Pictures", when(col("Pictures") == "No pictures available", 0)
    .otherwise(col("Pictures")))
display(df4)
```

```
from pyspark.sql.functions import transform, trim

df5 = df4.withColumn("Communication", transform(col("Communication"), lambda x: trim(x)))\
         .withColumn("Call", when(array_contains(col("Communication"), "Call"), True).otherwise(False))\
         .withColumn("Email", when(array_contains(col("Communication"), "Email"), True).otherwise(False))\
         .drop("Communication")
display(df5)
```

```
from pyspark.sql.functions import regexp_replace
from pyspark.sql.types import IntegerType

df6 = df5.withColumn("Area", col("BedBathandArea").getItem("Area"))\
         .withColumn("Area(Sq. Yd.)", regexp_replace("Area", " Sq. Yd.", "").cast(IntegerType()))\
         .withColumn("Bathrooms", df5.BedBathandArea.getItem("Baths").cast(IntegerType()))\
         .withColumn("Bedrooms", df5.BedBathandArea.getItem("Beds").cast(IntegerType()))\
         .drop("BedBathandArea", "Area")
display(df6)
```

```
df7 = df6.withColumn("AdDuration", regexp_replace("AdDuration", "Added: ", ""))
display(df7)
```

```
from pyspark.sql.functions import concat,format_string

df8 = df7.withColumn("AdURL", format_string("www.zameen.com%s", "AdURL"))
display(df8)
```

## Convert between DataFrames and SQL

```
df8.createOrReplaceTempView("ads1")
```

```
%sql

SELECT AdID, Title, `Area(Sq. Yd.)`, `Cost(Rs)` 
FROM ads1 WHERE `Area(Sq. Yd.)` <= 120
ORDER BY `Cost(Rs)` DESC;
```

```
df9 = df8.withColumnRenamed("Cost(Rs)", "Cost") \
    .withColumnRenamed("Area(Sq. Yd.)", "Area")
```
